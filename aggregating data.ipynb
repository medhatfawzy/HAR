```python
import pandas as pd
```

```python
X_train = pd.read_csv('data/train/X_train.txt', delim_whitespace=True, header=None)
X_train_columns = pd.read_csv('data/features.txt', delim_whitespace=True, header=None)
X_train_subject = pd.read_csv('data/train/subject_train.txt', delim_whitespace=True, header=None)
```

```python
X_train
```

```python
# Checking for duplication
X_train_columns[1].value_counts()
```

```python
X_train_subject
```

```python
X_train_columns = X_train_columns[1].tolist()
X_train.columns = X_train_columns
X_train['subject'] = X_train_subject
```

```python
X_train
```

```python
y_train = pd.read_csv('data/train/y_train.txt', delim_whitespace=True, header=None)
y_train.columns = ['activity']

y_train_labels = pd.read_csv('data/activity_labels.txt', delim_whitespace=True, header=None)
y_train_labels = y_train_labels[1].tolist()
```

```python
y_train.activity.value_counts()
```

```python
y_train_labels
```

```python
# replce the numbers with the labels
y_train.activity = y_train.activity.map(lambda x: y_train_labels[x-1])
```

```python
y_train.activity.value_counts()
```

```python
train_data = pd.concat([X_train, y_train], axis=1)
```

```python
train_data
```

```python
train_data.to_parquet('train_data.parquet')
```

```python
X_test = pd.read_csv('data/test/X_test.txt', delim_whitespace=True, header=None)
X_test.columns = X_train_columns
X_test_subject = pd.read_csv('data/test/subject_test.txt', delim_whitespace=True, header=None)

y_test = pd.read_csv('data/test/y_test.txt', delim_whitespace=True, header=None)
y_test.columns = ['activity']

y_test.activity = y_test.activity.map(lambda x: y_train_labels[x-1])
```

```python
X_test['subject'] = X_test_subject
X_test
```

```python
test_data = pd.concat([X_test, y_test], axis=1)
```

```python
test_data
```

```python
test_data.to_parquet('test_data.parquet')
```
